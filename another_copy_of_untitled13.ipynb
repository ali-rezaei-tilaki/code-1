```python
# SPT - Shortest Processing Time
jobs = {"Job1": 6, "Job2": 2, "Job3": 8, "Job4": 3}

# مرتب‌سازی بر اساس زمان پردازش (Shortest Processing Time)
spt_order = sorted(jobs.items(), key=lambda x: x[1])

waiting_time = 0
completion_time = 0
total_waiting = 0
total_completion = 0

print(f"{'Job':<6}{'Processing':<12}{'Waiting':<10}{'Completion':<12}")
for job, time in spt_order:
    completion_time = waiting_time + time
    print(f"{job:<6}{time:<12}{waiting_time:<10}{completion_time:<12}")
    total_waiting += waiting_time
    total_completion += completion_time
    waiting_time = completion_time

n = len(jobs)
print(f"\nAverage Waiting Time: {total_waiting / n:.2f}")
print(f"Average Completion Time: {total_completion / n:.2f}")
```

```python
# LPT - Largest Processing Time (پیشرفته)
jobs = {"Job1": 6, "Job2": 2, "Job3": 8, "Job4": 3}

lpt_order = sorted(jobs.items(), key=lambda x: x[1], reverse=True)

# محاسبه زمان انتظار و زمان پایان
waiting_time = 0
completion_time = 0
total_waiting = 0
total_completion = 0

print(f"{'Job':<6}{'Processing':<12}{'Waiting':<10}{'Completion':<12}")
for job, time in lpt_order:
    completion_time = waiting_time + time
    print(f"{job:<6}{time:<12}{waiting_time:<10}{completion_time:<12}")
    total_waiting += waiting_time
    total_completion += completion_time
    waiting_time = completion_time

n = len(jobs)
print(f"\nAverage Waiting Time: {total_waiting / n:.2f}")
print(f"Average Completion Time: {total_completion / n:.2f}")
```

```python
# RLPT - Reverse Largest Processing Time
jobs = {"Job1": 6, "Job2": 2, "Job3": 8, "Job4": 3}

# مرتب‌سازی بر اساس زمان پردازش (Reverse LPT: کوچک به بزرگ)
rlpt_order = sorted(jobs.items(), key=lambda x: x[1])


waiting_time = 0
completion_time = 0
total_waiting = 0
total_completion = 0

print(f"{'Job':<6}{'Processing':<12}{'Waiting':<10}{'Completion':<12}")
for job, time in rlpt_order:
    completion_time = waiting_time + time
    print(f"{job:<6}{time:<12}{waiting_time:<10}{completion_time:<12}")
    total_waiting += waiting_time
    total_completion += completion_time
    waiting_time = completion_time

n = len(jobs)
print(f"\nAverage Waiting Time: {total_waiting / n:.2f}")
print(f"Average Completion Time: {total_completion / n:.2f}")
```

```python
# LSPT - Largest Shortest Processing Time
jobs = {"Job1": 6, "Job2": 2, "Job3": 8, "Job4": 3}

# مرحله 1: مرتب‌سازی بر اساس بزرگترین زمان (LPT)
lpt_sorted = sorted(jobs.items(), key=lambda x: x[1], reverse=True)

# مرحله 2: داخل دسته‌ها (اینجا ساده سازی: همه جدا هستن)، می‌تونیم داخل دسته کوتاه‌ترین اول باشه
# برای مثال کوچک‌ترین زمان اول، که عملاً مثل SPT درون دسته‌ها
lsp_order = sorted(lpt_sorted, key=lambda x: x[1])

waiting_time = 0
completion_time = 0
total_waiting = 0
total_completion = 0

print(f"{'Job':<6}{'Processing':<12}{'Waiting':<10}{'Completion':<12}")
for job, time in lsp_order:
    completion_time = waiting_time + time
    print(f"{job:<6}{time:<12}{waiting_time:<10}{completion_time:<12}")
    total_waiting += waiting_time
    total_completion += completion_time
    waiting_time = completion_time

n = len(jobs)
print(f"\nAverage Waiting Time: {total_waiting / n:.2f}")
print(f"Average Completion Time: {total_completion / n:.2f}")
```
